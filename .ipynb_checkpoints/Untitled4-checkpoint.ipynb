```python
CACHED_FILE = "CACHED.json"

try:
    fcached  = open(CACHED_FILE,"r")
    read_cached = fcached.read()
    cached_json_loads = json.loads(read_cached)
    fcached.close()
except:
    cached_json_loads = {}

def params_unique_combination(baseurl, params_d):
    alphabetized_keys = sorted(params_d.keys())
    res = []
    for k in alphabetized_keys:
        return baseurl + "_".join(res)


def get_from_wiki(query_name):
    baseurl = "https://en.wikipedia.org/w/api.php"
    params_d = {}
    params_d["action"] = 'query'
    params_d["format"] = 'json'
    params_d["prop"]= "info|description|extracts|pageassessments|redirects"
    params_d['prop'] = 'revisions'
    params_d["rvslots"]= 'main'
    params_d["formatversion"]= "2"
    params_d["inprop"] = "url"
    params_d["titles"]= query_name
    params_d["rvprop"]= "timestamp|user|comment|content"
    params_d['rvlimit']='500'
    params_d["exlimit"] =  "1"
    params_d["redirects"]= "prop=redirects"
    unique_ident = params_unique_combination(baseurl,params_d)
    #resp = requests.get(baseurl, params_d)
    if unique_ident in cached_json_loads:
        return cached_json_loads[unique_ident]
    else:
        resp = requests.get(baseurl, params_d)
        cached_json_loads[unique_ident] = json.loads(resp.text)

        dumped_json_cache = json.dumps(cached_json_loads)

        fcached = open(CACHED_FILE,"w")
        fcached.write(dumped_json_cache)
        fcached.close()
        return cached_json_loads[unique_ident]
    time.sleep(2)   # Delays for 2 seconds.
    res_obj = requests.get(base_url, params_d)
    return json.loads(res_obj.text)


wikidata = get_from_wiki('Roger Federer')

print(wikidata)
```
