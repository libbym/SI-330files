### <font color="magenta">Q2: Read from the data from email_data_file_name using csv and DictReader and then store the data into email_data data structure.</font>

````python
#Solution block
email_data_file_name = "email_sample_5000.csv"

email_data = []

with open(email_data_file_name, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        email_data.append(row)
````

````python
bodies = [email['message'].split('\n\n')[1] for email in email_data]
````

````python
words = []
for text in bodies:
    words.extend( nltk.word_tokenize(text)) 
````

````python
words = []
for text in bodies:
    for word in (nltk.word_tokenize(text)):
        words.append( word.lower()) 
````

````python
words
````

````python
text = nltk.Text(words)
````

````python
text.concordance('accounting')
````

````python
text.similar('gift')
````

````python
tagged_text = nltk.pos_tag(text)
````
